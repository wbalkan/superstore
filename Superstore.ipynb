# Data Imputation
Fill in all data that can be inferred from other values in the row

```python
# fill in all row IDs
for row in range(len(df)):
    df.at[row, 'Row ID'] = row + 1.0
null_check()
```

```python
# uses values with matching fields in dependencies array to fill in empty target fields
def fill_fields(target, dependencies):
    all_fields = [target]
    for field in dependencies:
        all_fields.append(field)
    filled_df = df.dropna(subset=all_fields)
    # create dictionary mapping values of dependencies to values of target
    grouped_df = filled_df.groupby(dependencies)[target]
    dict_df = grouped_df.apply(lambda x: x.iloc[0])
    map_dict = dict_df.to_dict()
    def fill(row):
        if pd.notna(row[target]):
            return
        deps = []
        for dep in dependencies:
            deps.append(row[dep])
        if len(deps) > 1:
            deps = tuple(deps)
        else:
            deps = deps[0]
        if deps in map_dict:
            df.at[row.name, target] = map_dict[deps]
    df.apply(fill, axis=1)
    
```

```python
# dictionary mapping columns to the columns / groups of columns they depend on
# make some assumptions such as individual customers not making more than one order in a day and no customers from 
# same city/postal code have same name. Could remove these assumptions to impute less data
dependency_dict = {}
dependency_dict["Order ID"] = [
    ["Order Date", "Customer ID"],
    ["Ship Date", "Customer ID"],
    ["Order Date", "Customer Name"],
    ["Ship Date", "Customer Name"]
]
dependency_dict["Order Date"] = [
    ["Order ID"],
    ["Ship Date", "Customer ID"],
    ["Ship Date", "Customer Name"]
]
dependency_dict["Ship Date"] = [
    ["Order ID"],
    ["Order Date", "Customer ID"],
    ["Order Date", "Customer Name"]
]
dependency_dict["Ship Mode"] = [
    ["Order ID"],
    ["Order Date", "Customer ID"],
    ["Order Date", "Customer Name"],
    ["Ship Date", "Customer ID"],
    ["Ship Date", "Customer Name"]
]
dependency_dict["Customer ID"] = [
    ["Order ID"],
    ["Customer Name", "City"],
    ["Customer Name", "Postal Code"]
]
dependency_dict["Customer Name"] = [
    ["Order ID"],
    ["Customer ID"]
]
dependency_dict["Segment"] = [
    ["Order ID"],
    ["Customer ID"]
]
dependency_dict["Country"] = [
    ["Order ID"],
    ["Customer ID"],
    ["City"],
    ["State"],
    ["Postal Code"]
]
dependency_dict["City"] = [
    ["Order ID"],
    ["Customer ID"],
    ["Postal Code"]
]
dependency_dict["State"] = [
    ["Order ID"],
    ["Customer ID"],
    ["Postal Code"],
    ["City"]
]
dependency_dict["Postal Code"] = [
    ["Order ID"],
    ["Customer ID"],
]
dependency_dict["Region"] = [
    ["Order ID"],
    ["Customer ID"],
    ["Postal Code"],
    ["City"],
    ["State"]
]
dependency_dict["Product ID"] = [
    ["Product Name"]
]
dependency_dict["Category"] = [
    ["Product ID"],
    ["Sub-Category"],
    ["Product Name"]
]
dependency_dict["Sub-Category"] = [
    ["Product Name"],
    ["Product ID"]
]
dependency_dict["Product Name"] = [
    ["Product ID"]
]
```

```python

for i in range(2):
    for target in dependency_dict.keys():
        for deps in dependency_dict[target]:
            fill_fields(target, deps)

null_check()
```

```python
# fill out dictionaries that map product name & id to the price and cost of the product if necessary data is available
name_to_price = {}
id_to_price = {}
name_to_cost = {}
id_to_cost = {}
for index, row in df.iterrows():
    if pd.notna(row["Quantity"]) and pd.notna(row["Sales"]):
        if pd.notna(row["Discount"]):
            price = row["Sales"] / ( row["Quantity"] * (1 - row["Discount"]) )
            if pd.notna(row["Product ID"]):
                id_to_price[row["Product ID"]] = price
            if pd.notna(row["Product Name"]):
                name_to_price[row["Product Name"]] = price
        if pd.notna("Profit"):
            cost = (row["Sales"] - row["Profit"]) / row["Quantity"]
            if pd.notna(row["Product ID"]):
                id_to_cost[row["Product ID"]] = cost
            if pd.notna(row["Product Name"]):
                name_to_cost[row["Product Name"]] = cost
```

```python
# use price and quantity to fill out missing fields where possible
for index, row in df.iterrows():
    if row["Product Name"] in name_to_price:
        price = name_to_price[row["Product Name"]]
    elif row["Product ID"] in id_to_price:
        price = id_to_price[row["Product ID"]]
    else:
        price = pd.NA
    if row["Product Name"] in name_to_cost:
        cost = name_to_cost[row["Product Name"]]
    elif row["Product ID"] in id_to_cost:
        cost = id_to_cost[row["Product ID"]]
    else:
        cost = pd.NA
    if pd.isna(row["Sales"]):
        sales = price * row["Quantity"] *(1 - row["Discount"])
        if pd.notna(sales):
            df.loc[index, "Sales"] = sales
    if pd.isna(row["Quantity"]):
        quantity = row["Sales"] / (price * (1 - row["Discount"]))
        if pd.notna(quantity):
            df.loc[index, "Quantity"] = quantity
    if pd.isna(row["Discount"]):
        discount = 1 - ( row["Sales"] /(price * row["Quantity"]) )
        if pd.notna(discount):
            df.loc[index, "Discount"] = discount
    if pd.isna(row["Profit"]):
        profit = ( (price * (1 - row["Discount"])) - cost ) * row["Quantity"]
        if pd.notna(profit):
            df.loc[index, "Profit"] = profit
```

# Feature extraction

```python
# create a column for the total number of orders by a particular customer 
customer_group = clean_df.groupby(by="Customer ID")
num_orders = customer_group["Order ID"].count().reset_index()
num_orders.columns = ['Customer ID', 'Number of Orders']
num_orders
```

```python
# create a column for profit per order
profit_total = customer_group["Profit"].sum().reset_index()
profit_per = pd.merge(profit_total, num_orders, on='Customer ID', how='left')
profit_per['Profit per Order'] = profit_per['Profit'] / profit_per['Number of Orders']
profit_per.drop(columns='Profit', inplace=True)
profit_per
```

```python
df_counts = pd.merge(clean_df, profit_per, on='Customer ID', how='left')
df_counts
```
